```python
from clonedrepo.pornhub import PornHub
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
client = PornHub()
videos = []
for video in client.getVideos(100, page=1, sort_by="view", time="all", location="world"):
    videos.append(video)
df = pd.DataFrame(videos)
```

```python
fig, ax = plt.subplots(figsize=[20,5])
g = sns.barplot(df["views"], ax=ax);
# g.set_xticks()
g.set_xticks(ticks=range(100), labels=['#'+str(i+1) for i in range(100)], rotation=90);
g.set_yticks(ticks=range(0,int(3e8),int(5e7)), labels=range(0,300,50));
g.set_xlabel('Video rank');
g.set_ylabel('Views (in millions)');
# g.set_xticklabels(df["title"], rotation=90)
plt.plot(range(100), [df["views"][0]/(x+1) for x in range(100)], 'r--', label='1/x');
plt.title("Do Pornhub video views obey zipf's law?");
plt.legend(loc='best');
plt.savefig('NoItDoesNot.png')
```
